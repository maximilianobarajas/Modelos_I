#### Listas

Las listas son el objeto que usaremos en Sage para trabajar con conjuntos. Las listas no son del todo conjuntos, pues permiten repeticiones y el orden importa, pero son suficientemente parecidas y son fáciles de manejar. Aqui hay un ejemplo de una lista.



```python
Primos=[2,3,5,7,11,13]
Primos
```

Las listas se manejan de manera parecida a los arreglos en la programación, usamos corchetes para definirlas y listamos sus elementos usando comas. Para referirnos a un elemento de la lista usaremos los corchetes y el índice de la posición del elemento. Los índices empiezan en cero.

```python
Primos[3]
```

Podemos agregar un elemento a la lista al final de manera muy sencilla con el método `append`

```python
Primos.append(17); Primos
```

Tambien podemos quitarle un elemento con el método `remove`

```python
Primos.remove(11);Primos
```

Tambien podemos extraer una sublista, como puedes ver en el siguiente ejemplo:

```python
ImparesChicos= Primos[1:4]; ImparesChicos
```

Otra cosa que podemos hacer es crear una lista por comprensión, es decir definir una nueva lista a partir de otra modificando sus elementos o aplicando una función a los mismos.

```python
Sucesores=[n+1 for n in Primos];Sucesores
```

En el ejemplo anterior a cada $n$ en la lista `Primos`le aplicamos la función $F(n) = n+1$, es decir le sumamos un elemento y lo guardamos en una nueva lista que se llama Sucesores. Existen algunas funciones que nos ayudan a generar listas de números enteros de manera eficiente. La función `srange(n)` nos permite generar una lista con los números del $0$ al $n-1$, como se puede ver en la siguiente celda.

```python
Lista=srange(10);Lista
```

Si en lugar de darle $n$ a la función `srange()` como argumento, le damos dos valores $n$ y $m$, es decir ponemos `srange(n,m)` obtendremos una lista de números que empieza en el $n$ y acaba en el $m-1$, como puedes ver en el siguiente ejemplo.

```python
Adolescencia= srange(13,20); Adolescencia
```

Por último si le damos como argumentos a la función `srange()` los valores de $n$, $m$, y $k$, es decir ponemos  `srange(n,m,k)`obtendremos una lista de enteros que empieza en el $n$, y va avanzando de $k$ en $k$ mientras el número sea menor que $m$, como puedes ver en la siguiente celda.

```python
Bisiestos=srange(1960, 2000,4);Bisiestos
```

Observa que en el ejemplo anterior, aun cuando $1996+4=2000$ no aparece en la lista, pues solo pone todos los números que empiezan en 1960 y se les va sumando de 4 en 4  y sean menores que 2000.

#### Graficas de funciones

La función `plot()` nos permite graficar funciones, su rango por default es $-1 \leq x \leq 1$. A veces aparece un mensaje de error, pero puedes ignorarlo, solo esta cargando las librerias necesarias.


```python
plot(x^2)
```

```python
plot (x^3-x)
```

```python
plot(abs(x-1/2))
```

Si queremos podemos especificar el dominio de la función:

```python
plot(x^3-x, -2, 2)
```

```python
plot(x^4 - 3*x^2+2,-2,2)
```

```python
plot(sin(x),-10,10)
```

Podemos quitar los ejes:

```python
plot(x^3, 5, 10, axes=False)
```

Tambien podemos acotar los valores en el rango de $y$

```python
plot( x^3-x, -3, 3, ymin = -6, ymax = 6)
```

Esto es particularmente útil cuando tenemos asíntotas

```python
plot(1/(x^3-x), -2, 2)
```

```python
plot(1/(x^3-x), -2, 2, ymin = -5, ymax = 5)
```

```python
plot(1/(x^3-x), (x,-2, 2), detect_poles=True, ymin = -5, ymax = 5)
```

```python
plot(1/(x^3-x), (x,-2, 2), detect_poles='show', ymin = -5, ymax = 5)
```

Como puedes ver en la primera gráfica no se ve nada, esto es por que sage intenta que aparezcan todos los puntos y como hay una asíntota la funcion crece demasiado, la escala sobre el eje $y$ se hace muy pequeña y se pierde la imagen. El segundo ejemplo corta los valores muy grandes y muy pequeños de la gráfica, pero intenta conectar las partes. En el tercer ejemplo deja de intentar conectar estas partes y en el cuarto ya muestra las asíntotas.  Podemos tambien superponer gráficas, por ejemplo una parábola y la recta tangente a un punto de ella. Para superponerlas usaremos el signo +

```python
plot( 2*x, -2, 2 ) + plot( x^2+1, -2, 2 )
```

Tambien podemos agregar el punto en el que se intersectan y una rejilla para que se vea aun mejor

```python
plot( 2*x, -2, 2, gridlines='minor' ) + plot( x^2+1, -2, 2 ) + point( (1,2), size=30, color="red" )
```

Y podemos hacer con múltiples gráficas

```python
plot((x-1)*(x-2)*(x-3), -2, 10) + plot((x-1)*(x-2)*(x-4), -2, 10) + plot((x-1)*(x-2)*(x-5), -2,10) + plot((x-1)*(x-2)*(x-6), -2, 10) + plot((x-1)*(x-2)*(x-7), -2, 10)
```

Aunque como puedes ver eso puede empezar a hacer el código ilegible. Par que sea mas entendible podemos usar unas variables auxiliares:

```python
P1 = plot( (x-1)*(x-2)*(x-3), 1, 6)
P2 = plot( (x-1)*(x-2)*(x-4), 1, 6)
P3 = plot( (x-1)*(x-2)*(x-5), 1, 6)
P4 = plot( (x-1)*(x-2)*(x-6), 1, 6)
P5 = plot( (x-1)*(x-2)*(x-7), 1, 6)
P = P1 + P2 + P3 + P4 + P5
P.show()
```

Por último a veces es conveniente cambiar de colores

Una familia de funciones es un conjunto de funciones. Usualmente están definidas por uno o más parámetros. Para esta práctica deberas hacer un dibujo con una familia de funciones donde aparezcan al menos 10 elementos de la familia con distintos parámetros. 

La familia que decidí usar fue la familia de funciones de la distribución Gamma, recordemos que la función de densidad de probabilidad de la función Gamma se ve definida como sigue:

$$f(x;\alpha,\beta) = \frac{\beta^\alpha x^{\alpha-1} e^{-\beta x}}{\Gamma(\alpha)}$$

Implemente la función como a continuación:

```python
def Gamma_FDP(x: float, alpha: float, beta: float):
    return beta^alpha * x^(alpha-1) * exp(-beta * x) / gamma(alpha)
```

```python
Gamma_FDP(2,3,2.1)
```
